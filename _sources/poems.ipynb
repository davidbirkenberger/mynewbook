```python
import pandas as pd
```

```python
df = pd.read_csv(r'K:\python\poems\POetryFoundationData.csv', index_col=0)
df.head()
```

```python
Poets = df["Poet"].to_list()
Poets
```

```python
from bs4 import BeautifulSoup
import requests
import json

metadata = []
for i in df["Poet"]:
    Poet = Poets[i]
    Poet_lower = Poet.lower().split()
    final_poet = Poet_lower[0] + "-" + Poet_lower[1]
    base_url = "https://www.poetryfoundation.org/poets/"
    final_url = base_url + final_poet
    source_code = requests.get(final_url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text)
    birthday_raw = soup.find("span", {"class":"c-txt c-txt_poetMeta"})
    print(birthday_raw)
   
 
```
